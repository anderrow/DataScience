#Get Data


```python
import os
import pandas as pd
import numpy as np

path = "AmaboxData/UfaProd.csv"
df = pd.read_csv(path)
```

# Data frame construction and cleaning

```python
df.head(3) #Visualisation
```

```python
df.set_index('Id', inplace=True) #Set Id as Index
```

```python
df.duplicated().any() #Any duplicate values?
df.duplicated().unique()
df.columns #Check valuables columns
```

```python
df = df.drop(columns=[col for col in df.columns if 'Unnamed' in col]) #Erase all 'Unnamed colums'
df.columns #Check again for valuables columns
df.sample(1)
```

```python
df_prop = df[['RequestedAmount', 'ArticleDensity','LowerProportioningTolerance','UpperProportioningTolerance', 'Repaired','ActualAmount','Name']]
df_prop.rename(columns={'Name': 'Ingredient'}, inplace=True)
df_prop.sample(1)
```

```python
df_prop.info() #Check if Dtype is correct
```

```python
df_prop.isna().any() #Check for Nan Values
```

```python
df_prop.describe()
```

```python
df_prop[df_prop['UpperProportioningTolerance'] == 0]
df_prop = df_prop[df_prop['UpperProportioningTolerance'] != 0] #Keep only the lines that are not conversions
```

```python
len(df_prop[df_prop['UpperProportioningTolerance'] > 15]) == len(df_prop[df_prop['UpperProportioningTolerance'] == 20]) #Must be a bug in which one of the tolerances goes to the top
```

```python
df_prop = df_prop[df_prop['UpperProportioningTolerance'] != 20] #Keep only the lines whose tolerance is not 20
df_prop.describe() #Check that the values are now more reasonable
```

```python
#df_prop_intake = df_prop[df_prop['RequestedAmount'] == -1] #Intakes --> Unseful DF needs more info. Requested Amount is needed.Lower and Upper Proportioning for the Intakes also
df_prop_doser = df_prop[df_prop['RequestedAmount'] != -1] #Proportionings
#df_prop_intake.describe() #Unseful DF needs more info. Requested Amount is needed.Lower and Upper Proportioning for the Intakes also
```

```python
df_prop_doser.describe()
```

```python
df_prop_doser['Gap'] = df_prop_doser['ActualAmount'] - df_prop_doser['RequestedAmount']
```

```python
df_prop_doser.describe()
```

## Proportionings out of tolerance at the doser

```python
df_prop_doser_outoftol= df_prop_doser[(df_prop_doser['RequestedAmount'] - df_prop_doser['LowerProportioningTolerance'] > df_prop_doser['ActualAmount']) &
              (df_prop_doser['RequestedAmount'] + df_prop_doser['UpperProportioningTolerance'] < df_prop_doser['ActualAmount'])]
```
